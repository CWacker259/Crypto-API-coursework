## Investment Multiple and CAGR  

**Two alternative reward metrics that are more intuitive and easier to interpret**

```python
msft = pd.read_csv('msft.csv', index_col = "Date", parse_dates = ["Date"])
```

**Investment Multiple**: Ending value of 1 (dollar) invested. 
multiple = Ending value / initial investment

```python
multiple = (msft.Price[-1] / msft.Price[0]) 
print("multiple: {}".format(multiple))
```

**Price Increase (in %)** 

```python
PI = (multiple - 1) * 100 
print('Price Increase: {}%'.format(PI))
```

```python
msft.Price / msft.Price[0] #similar/indentical concept: Normalized Price with Base Value 1 
```

**Drawback of investment multiple:** Doesn't take into account investment Period. Meaningful only in conjunction with investment period 

**Compound Annual Growth Rate (CAGR):** the (constant annual) rate of return that would be required for an investment to grow from its beginning  balance to its ending balance, assuming the profits were reinvested at the end of each year of the investment's lifespan

```python
start = msft.index[0] 
start
```

```python
end = msft.index[-1] 
end
```

```python
td = end - start 
td
```

```python
td_years = td.days / 365.25 
td_years
```

```python
cagr = multiple**(1 / td_years) - 1 #short version 
cagr
```

```python
cagr = (msft.Price[-1]/msft.Price[0])**(1/((msft.index[-1] - msft.index[0]).days / 365.25)) - 1 #long version 
cagr
```

```python
(1 + cagr)**(td_years) # alternative #2 to calculate multiple (cagr)
```

**CAGR can be used to compare investments with different investment horizons**

## Discrete Compounding 

**Annual Compounding** -> interests accrue once a years at the end of the year 

your Savings Bank offers an interest rate of **8% p.a. (stated rate)** with **annual compounding** on your savings ($100) 
Calculate the **value** of your savings account after one year and the corresponding **effective annual interest rate**

**Interests are calculated and added to your savings account once at the end of each year**

```python
PV = 100 
r = 0.08 
n = 1 
```

```python
FV = PV * (1 + r)**n 
FV
```

```python
effective_annual_rate = (FV / PV)**(1/n) - 1 
effective_annual_rate
```

Quarterly Compounding, interests accrue once per quarter at the end of the quarter

```python
PV = 100 
r = 0.08 
n = 1 
m = 4 
```

```python
FV = PV * (1 + r/m)**(n*m) 
FV
```

```python
effective_annual_rate = (FV / PV)**(1/n) - 1 
effective_annual_rate
```

Monthly compounding 

```python
PV = 100 
r = 0.08 
n = 1 
m = 12 
```

```python
FV = PV * (1 + r/m)**(n*m) 
FV
```

```python
effective_annual_rate = (FV / PV)**(1/n) - 1 
effective_annual_rate
```

## Continuous Compounding 

your Savings Bank offers an interest rate of **8% p.a. (stated rate)** with **annual compounding** on your savings ($100) 
Calculate the **value** of your savings account after one year and the corresponding **effective annual interest rate**

```python
PV = 100 
r = 0.08 
n = 1 
m = 10000000 #approx infinity 
```

```python
FV = PV * (1 + r/m)**(n*m) #approx with large m
FV
```

```python
FV = PV * np.exp(n * r) #exact math with e  
FV
```

```python
euler = np.exp(1) 
euler
```

```python
PV * euler**(n * r)
```

```python
effective_annual_rate = np.exp(r) - 1 
effective_annual_rate
```

Lets assume we only observe PV and FV, how to calculate the stated rate/continuously compounded rate of 8%?

```python
r = np.log(FV / PV) #inverse calculation using log
r
```

intuitively, it makes a lot of sense to work with log returns**

```python
 100 * (1 + geo_mean)**periods # CORRECT 
```

```python
sum_lr = df.LR.sum() 
sum_lr
```

```python
100 * np.exp(sum_lr) # correct 
```

```python
mean_lr = df.LR.mean() 
mean_lr
```

**Log returns are additive over time. Simple returns are not additive over time( but the can be multipled/compounded)**

## Normality of Financial Returns 

```python
msft = pd.read_csv('msft.csv', index_col = "Date", parse_dates = ["Date"])  
```

```python
msft.describe()
```

```python
msft.log_ret.plot(kind = 'hist',figsize = (15,8),bins = 100,fontsize=15,density = False) #Frequency Distribution of log returns
plt.xlabel('Daily Returns',fontsize = 15)                             #is density is set to true the area of graph will equal 1
plt.ylabel('Frequency', fontsize = 15) 
plt.title('Frequency Distribution of Returns',fontsize = 20) 
plt.show()
```

Higher Central Moments are zero: 

* skew = 0  
* (Excess) Kurtosis = 0 (positive excess Kurtosis -> more observations in the 'tails')

```python
mu = msft.log_ret.mean() 
sigma = msft.log_ret.std() 
print(mu) 
print(sigma)
```

```python
import scipy.stats as stats 
```

```python
stats.skew(msft.log_ret.dropna()) # 0 means its symmetric, so msft is slightly skewed to the left
```

```python
stats.kurtosis(msft.log_ret.dropna(), fisher = True) #in a normal distribution: (fisher) kurtosis == 0
```

**MSFT Returns exhibit "Fat Tails" (extreme positive/negative outcomes)**

```python
x = np.linspace(msft.log_ret.min(), msft.log_ret.max(), 10000) #10000 x values between the log_ret min and max
x
```

```python
y = stats.norm.pdf(x, loc = mu, scale = sigma) # creating y values for a normal distribution with mu, sigma 
y
```

```python
plt.figure(figsize = (20,8)) 
plt.hist(msft.log_ret, bins = 500, density = True, label = "Frequency Distribution of daily Returns (MSFT)") 
plt.plot(x, y, linewidth = 3, color= 'red', label = 'Normal Distribution')
plt.title('Normal Distribution', fontsize = 20)
plt.xlabel('Daily Returns', fontsize =15) 
plt.ylabel('pdf', fontsize = 15) 
plt.legend(fontsize = 15) 
plt.show()
```

**Testing the normality of MSFT Returns based on the sample given**  
-> Hypothesis test with H0 hypothesis: MSFT returns (full population) follow a normal Distribution 

```python
z_stat, p_value = stats.normaltest(msft.log_ret.dropna()) 
```

```python
z_stat #high values -> reject H0 
```

```python
p_value #low values (close to 0) reject h0
```

```python
round(p_value, 10)
```

MSFT returns don't follow a normal distribution as they exhibit 'fat tails'.

### How will the Mean-Variance Analysis change with smoothed data? 

```python
freqs = ['A','Q','M','W-Fri','D']
periods = [1,4,12,52,252]
ann_mean = []
ann_std = []
```

```python
for i in range(5): 
    resamp = msft.Price.resample(freqs[i]).last() #resample
    ann_mean.append(np.log(resamp / resamp.shift()).mean() * periods[i]) #calc. annualized mean 
    ann_std.append(np.log(resamp / resamp.shift()).std() * np.sqrt(periods[i])) #calc annualized std
```

```python
summary = pd.DataFrame(data = {'ann_std':ann_std, 'ann_mean':ann_mean}, index = freqs) 
summary
```

```python
summary.plot(kind = 'scatter', x='ann_std', y='ann_mean', figsize=(15,8), s = 50, fontsize=15) 
for i in summary.index: 
    plt.annotate(i, xy=(summary.loc[i, 'ann_std']+0.001, summary.loc[i, 'ann_mean']+0.001), size=15) 
plt.ylim(0, 0.3) 
plt.xlabel('ann. Risk(std)',fontsize=15) 
plt.ylabel('ann. Return', fontsize=15) 
plt.title('Risk/Return', fontsize=20) 
plt.show() 
```

**-> Smoothing reduces (observed) risk** 

Dubious practices: 
* managing (manipulating) performance in performance reportings

* comparing assets with different pricing frequency and pricing mechanisms -> e.g. real estate with quarterly valuation cs. listed stocks (minutely/hourly/daily prices)

## Margin Trading & Levered Returns 

**Definition:** Margin trading refers to the practive of using borrowed funds from a broker to trade a financial asset, which forms the collateral for the loan from the broker 

In simple words: investors **don't pay the full price** but they get the full benefit (less borrowing costs) 

Its a two edged sword: leverage **amplifies both gains and losses** 

In the event of a loss, the collateral gets reduced and the investor either posts additional margin or the brokers closes the position 

**Example** 

A trader buys a stock (stock price: 100) **on margin (50%)**. After one day the prices increases to 110. 

Calculate **unlevered return** and **levered return**

```python
P0 = 100 
P1 = 110 
leverage = 2 
margin = P0/2  
print(margin)
```

```python
unlev_return = (P1 - P0) / P0 #simple return   
unlev_return 
```

```python
lev_return = (P1 - P0) / margin #simple return 
lev_return
```

```python
lev_return == unlev_return * leverage #this relationship is true for simple returns 
```

```python
unlev_return = np.log((P1 - P0) / P0 + 1) #log return 
unlev_return
```

```python
lev_return = np.log((P1 - P0) / margin + 1) #log return 
lev_return
```

```python
lev_return == unlev_return * leverage # this relationship does not hold true for log returns 
```

**Take Home: to calculate levered returns, don't multiply leverage with log returns**
