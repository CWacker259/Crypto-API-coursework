1. Calculate daily returns for Bitcoin.

2. Calculate the arithmetic mean return and the standard deviation of returns for Bitcoin.

3. Compare Bitcoin with Microsoft (mu = 0.00116, sigma = 0.01726). Does the rule "Higher Risk -> Higher Reward" hold?

```python
import pandas as pd 
import yfinance as yf  
import matplotlib.pyplot as plt 
import numpy as np
pd.options.display.float_format = '{:.4f}'.format
plt.style.use('seaborn')
```

```python
start = '2014-10-01' 
end = '2021-05-21' 
symbol = ['MSFT','BTC-USD'] 
df = yf.download(symbol,start,end) 
```

```python
df = df.Close
```

```python
msft = df.MSFT.dropna().to_frame().copy() 
```

```python
btc = df['BTC-USD'].dropna().to_frame().copy()
```

```python
msft.rename(columns={'MSFT':'Price'},inplace = True)
msft['Returns'] = msft.Price.pct_change(periods = 1) 
msft['Returns'].to_frame().describe()
```

```python
btc.rename(columns = {'BTC-USD':'Price'}, inplace = True) 
btc['Returns'] = btc.Price.pct_change(periods = 1) 
btc['Returns'].to_frame().describe()
```

```python
print("Microsoft AVG return: {}".format(msft.Returns.mean())) 
print("BTC AVG return: {}".format(btc.Returns.mean()))
print("Microsoft Volatility: {}".format(msft.Returns.std()))  
print("BTC Volatility: {}".format(btc.Returns.std())) 
```
