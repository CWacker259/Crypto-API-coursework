```python
import pandas as pd 
import yfinance as yf 
import matplotlib.pyplot as plt 
pd.options.display.float_format = '{:.4f}'.format
plt.style.use('seaborn')
```

```python
start = '2015-01-02' 
end = '2020-12-31'
symbol = ['GE','TSLA'] 
```

```python
df = yf.download(symbol, start, end) 
```

```python
df
```

```python
close = df.Close.copy()
```

```python
close 
```

```python
norm = close.div(close.iloc[0]).mul(100) 
norm
```
